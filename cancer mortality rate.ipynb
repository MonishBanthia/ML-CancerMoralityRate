```python
# this function will return any column which contains single values

def find_constaint_colums(dataframe):

  constant_columns = []
  for column in dataframe.columns:

    unique_values = dataframe[column].unique()

    if len(unique_values) ==1:
      constant_columns.append(column)
  return constant_columns

# print(constant_columns)
```

```python
# this funtion returns the few unique values in the column

def find_column_with_few(dataframe,threshold):
  few_columns = []
  for column in dataframe.columns:

    unique_values = dataframe[column].unique()

    if len(unique_values) < threshold:
      few_columns.append(column)
  return few_columns
```

```python
def bin_to_num(data):
  binnedinc = []
  for i in data['binnedInc']:
    # this strip fx will remove the () and [] from the binnedInc column
    i = i.strip("()[]")
    # split the string into list
    i = i.split(",")
    # convert the list to tuple
    i = tuple(i)
    i = tuple(map(float,i))
    i = list(i)

    binnedinc.append(i)

  data['binnedInc'] = binnedinc


  data['lower_bound'] = [i[0] for i in data['binnedInc']]
  data['upper_bound'] = [i[1] for i in data['binnedInc']]

  data['median'] = (data['lower_bound'] + data['upper_bound']) / 2

  data.drop('binnedInc', axis = 1 ,inplace=True)
  return data



```

```python
# the model is not going to understand the english stuff it should be numeric
# this function will convert the letters into numeric code so that the machine will understand it easily

from sklearn.preprocessing import OneHotEncoder

def one_hot_encoding(X):
  # select categorical columns
  categorical_columns = X.select_dtypes(include=['object']).columns

  one_hot_encoder = OneHotEncoder(sparse = False , handle_unknown='ignore')
  one_hot_encoded = one_hot_encoder.fit_transform(X[categorical_columns])

  one_hot_encoded = pd.DataFrame(
      one_hot_encoded , columns = one_hot_encoder.get_feature_names_out(categorical_columns)
  )

  X = X.drop(categorical_columns, axis = 1)
  X = pd.concat([X ,one_hot_encoded] , axis =1)
  return X
```

```python
# filling and droping the missing values and column

def drop_and_fill(dataframe):
  # get the columns with greater 50% missing values
  cols_to_drop = dataframe.columns[dataframe.isnull().mean() > 0.5]
  # drop the column
  dataframe = dataframe.drop(cols_to_drop , axis = 1)

  # fill the missing values with mean if missing values in column mean in less than 50%
  dataframe = dataframe.fillna(dataframe.mean())
  return dataframe

```

```python
data = drop_and_fill(data)
```

```python
# using z score for outliers detection
from scipy.stats import normaltest

numerical_cols = processed_data.select_dtypes(include = np.number).columns

normal_dist = []
non_normal_dist = []

for col in numerical_cols:
  stat ,p = normaltest(processed_data[col])
  print("statistics = %.3f , p = %.3f" % (stat,p))
  alpha = 0.05
  if p > alpha:
    normal_dist.append(col)
    # processed_data['normal_dist'] = normal_dist
    # aka we reject null hypo.
  else:
    non_normal_dist.append(col)
    # processed_data['non_normal_dist'] = non_normal_dist
print(normal_dist)
```

```python
def deal_with_outliers(df,col,basic_info):
  highest_allowed = basic_info[col]['mean'] + 3*basic_info[col]['std']
  lowest_allowed = basic_info[col]['mean'] - 3*basic_info[col]['std']
  df = df[(df[col]> highest_allowed) | (df[col]< lowest_allowed)]
  return df


cols_have_outliers = []
for col in normal_dist:
  df = deal_with_outliers(normal_data,col,basic_info_of_normal)
  shape = df.shape
  if shape[0]>0:
    cols_have_outliers.append(col)
print(cols_have_outliers)
```

```python
def iden_for_skew(df,cols):
  skewed_cols = []
  for col in cols:
    skew = df[col].skew()
    if(skew > 1 or skew < -1):
      skewed_cols.append(col)
  return skewed_cols

skewed_cols = iden_for_skew(data_for_skewness,data_for_skewness.columns)
len(skewed_cols)
```

```python
def box_plot(df,col):
  fig = go.Figure(data=[go.Box(y=df[col],boxpoints = 'outliers',jitter=0.3,pointpos=-1.8)])
  fig.update_layout(
      title= 'Boxplot of '+ col,
      yaxis_title=col,
      width=500,
      height=500)
  fig.show()

for col in skewed_cols:
  box_plot(processed_data,col)
```

```python
def deal_with_outliers_with_iqr(df,col):
  per25 = df[col].quantile(0.25)
  per75 = df[col].quantile(0.75)

  iqr = per75 - per25
  upper_limit = per75 + 1.5*iqr
  lower_limit = per25 - 1.5*iqr

  outliers = df[(df[col] > upper_limit) | (df[col] < lower_limit)]
  return outliers

cols_with_outliers = []
for col in skewed_cols:
  outliers = deal_with_outliers_with_iqr(processed_data,col)
  shape = outliers.shape
  if shape[0]>0:
    cols_with_outliers.append(col)
print(cols_with_outliers)
```

```python
# from the outliers from data with capping
def capping(df,col):
  per25 = df[col].quantile(0.25)
  per75 = df[col].quantile(0.75)

  iqr = per75 - per25
  upper_limit = per75 + 1.5*iqr
  lower_limit = per25 - 1.5*iqr

  df.loc[df[col] > upper_limit,col] = upper_limit
  df.loc[df[col] < lower_limit,col] = lower_limit
  return df

for col in cols_with_outliers:
  capped_data = capping(processed_data,col)
capped_data.shape


```

```python
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm

def corr_among_numeric_features(df,cols):
  numeric_col = df[cols]
  corr = numeric_col.corr()

  corr_features = set()
  for i in range(len(corr.columns)):
    for j in range(i):
      if abs(corr.iloc[i,j])>0.8:
        colname = corr.columns[i]
        corr_features.add(colname)
  return corr_features


def lr_model(x_train , y_train):
  x_train_with_intercept = sm.add_constant(x_train)
  lr = sm.OLS(y_train,x_train_with_intercept).fit()
  return lr


corr_features = corr_among_numeric_features(capped_data,capped_data.columns)
print(corr_features)
```
